- 면적이 130 이상이고 가격이 1.5억 미만인 아파트의 가격만 출력

```python
df2 = df[df.면적 >= 130]
```

```python
df2.loc[df2.가격 < 15000, ['가격']]
```

```python
df.loc[(df['면적'] >=130) & (df['가격'] <15000), ['가격']]
```

- 위의 3가지 조건을 모두 만족하는 아파트들의 가격의 합계를 출력

```python
df.loc[(df['면적'] >=130) & (df['가격'] <15000), '가격'].sum()
```
